```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/2024matches.csv")
```

```python
### Initial Observations
#- The dataset contains match-level data for ATP matches from the 2024 season
#- Each row represents a single match, with winner and loser information stored in separate columns
#- The dataset does not contain bookmaker odds; odds data will be sourced separately and merged at a later stage
#- Player rankings and other performance-related variables contain missing values
#- Match context variables such as surface and tournament level are available
```

```python
### Feature Considerations
#- The dataset contains both pre-match contextual variables and post-match performance statistics
#- Only variables known prior to match start (e.g. rankings, surface, tournament level) are suitable for probability estimation
# Post-match statistics (e.g. aces, break points, match duration) will be excluded to avoid data leakage
#- Match outcomes will be restructured from a winner/loser format into a player/opponent format for modelling
```

```python
# Pre-match contextual columns only
base_cols = [
    "tourney_date", "surface", "tourney_level", "round", "best_of",
    
    "winner_id", "winner_name", "winner_age", "winner_ht",
    "winner_hand", "winner_rank", "winner_rank_points",
    
    "loser_id", "loser_name", "loser_age", "loser_ht",
    "loser_hand", "loser_rank", "loser_rank_points"
]

df_base = df[base_cols].copy()
```

```python
winner_view = df_base.rename(columns={
    "winner_id": "player_id",
    "winner_name": "player_name",
    "winner_age": "player_age",
    "winner_ht": "player_ht",
    "winner_hand": "player_hand",
    "winner_rank": "player_rank",
    "winner_rank_points": "player_rank_points",
    
    "loser_id": "opp_id",
    "loser_name": "opp_name",
    "loser_age": "opp_age",
    "loser_ht": "opp_ht",
    "loser_hand": "opp_hand",
    "loser_rank": "opp_rank",
    "loser_rank_points": "opp_rank_points"
})

winner_view["win"] = 1
```

```python
loser_view = df_base.rename(columns={
    "loser_id": "player_id",
    "loser_name": "player_name",
    "loser_age": "player_age",
    "loser_ht": "player_ht",
    "loser_hand": "player_hand",
    "loser_rank": "player_rank",
    "loser_rank_points": "player_rank_points",
    
    "winner_id": "opp_id",
    "winner_name": "opp_name",
    "winner_age": "opp_age",
    "winner_ht": "opp_ht",
    "winner_hand": "opp_hand",
    "winner_rank": "opp_rank",
    "winner_rank_points": "opp_rank_points"
})

loser_view["win"] = 0
```

```python
df_long = pd.concat([winner_view, loser_view], ignore_index=True)
```

```python
df_long.shape
```

```python
df_long["win"].value_counts(normalize=True)
```

```python
df_long["rank_diff"] = df_long["player_rank"] - df_long["opp_rank"]
```

```python
df_long["rank_points_diff"] = (
    df_long["player_rank_points"] - df_long["opp_rank_points"]
)
```

```python
df_long["age_diff"] = df_long["player_age"] - df_long["opp_age"]
```

```python
df_long["height_diff"] = df_long["player_ht"] - df_long["opp_ht"]
```

```python
df_model = df_long.dropna(
    subset=["rank_diff", "rank_points_diff", "age_diff"]
).copy()
```

```python
df_model[["rank_diff", "age_diff"]].describe()
```

```python
### Feature Validation
#Summary statistics for engineered features indicate symmetric distributions centred around zero, 
#as expected given the player–opponent construction. Ranking differences show a wide spread, reflecting variation in competitive matchups, 
#while age differences are more tightly distributed. No implausible values were observed, 
#suggesting the engineered features are suitable for probabilistic modelling.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
features = [
    "rank_diff",
    "rank_points_diff",
    "age_diff"
]

X = df_model[features]
y = df_model["win"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_test_proba = model.predict_proba(X_test)[:, 1]
```

```python
pd.Series(y_test_proba).describe()
```

```python
coef_df = pd.DataFrame({
    "feature": features,
    "coefficient": model.coef_[0]
}).sort_values(by="coefficient")

coef_df
```

```python
eval_df = pd.DataFrame({
    "predicted_proba": y_test_proba,
    "actual_win": y_test.values
})
```

```python
eval_df["proba_bin"] = pd.cut(
    eval_df["predicted_proba"],
    bins=np.linspace(0, 1, 11)
)
```

```python
calibration_df = (
    eval_df
    .groupby("proba_bin")
    .agg(
        mean_predicted_proba=("predicted_proba", "mean"),
        observed_win_rate=("actual_win", "mean"),
        count=("actual_win", "size")
    )
    .reset_index()
)

calibration_df
```

```python
plt.figure()
plt.plot(
    calibration_df["mean_predicted_proba"],
    calibration_df["observed_win_rate"],
    marker="o",
    label="Model"
)

plt.plot([0, 1], [0, 1], linestyle="--", label="Perfect calibration")

plt.xlabel("Predicted Probability")
plt.ylabel("Observed Win Rate")
plt.title("Calibration Curve – Baseline Logistic Regression")
plt.legend()
plt.show()
```

```python
### Baseline Probability Model

#A logistic regression model was trained to estimate match win probabilities using relative ranking and age differences. 
    #Calibration analysis on held-out data indicates that the model is reasonably well calibrated across most probability ranges, 
    #with deviations at extreme probabilities likely driven by smaller sample sizes. 
    #This provides a transparent baseline for evaluating pricing accuracy in professional tennis matches.
```

```python
#Merging datasets has not gone as intended, given these results. Will now use a side-based calibration instead 
```

```python
market_df = pd.concat([
    odds_df.assign(
        implied_proba=1 / odds_df["B365W"],
        win=1
    )[["implied_proba", "win"]],
    
    odds_df.assign(
        implied_proba=1 / odds_df["B365L"],
        win=0
    )[["implied_proba", "win"]]
], ignore_index=True)
```

```python
market_df.groupby("win")["implied_proba"].mean()
```

```python
market_df["proba_bin"] = pd.cut(
    market_df["implied_proba"],
    bins=np.linspace(0, 1, 11)
)

market_calibration = (
    market_df
    .groupby("proba_bin")
    .agg(
        mean_predicted=("implied_proba", "mean"),
        observed_win_rate=("win", "mean"),
        count=("win", "size")
    )
    .reset_index()
)

market_calibration
```

```python
## ### Calibration Comparison: Data-Driven Model vs Betting Market

##The calibration plot compares predicted win probabilities from a logistic regression model against implied probabilities derived from Bet365 odds. 
##Both approaches exhibit strong overall calibration, with observed win rates closely tracking predicted probabilities across most probability bins.

##The data-driven model performs particularly well in mid-range probability bands (approximately 0.3–0.7), 
##where its calibration is comparable to — and in some cases marginally closer to — perfect calibration than the bookmaker market.
##At probability extremes, especially for low-probability outcomes, both the model and the market display increased variance, 
##reflecting smaller sample sizes and inherent uncertainty.

##Despite relying solely on publicly available historical features such as player rankings, ranking points, and age differences, 
## The model captures a substantial portion of the probabilistic structure embedded in market odds. 
##This suggests that historical performance data contains a meaningful predictive signal consistent with professional pricing, 
##while also highlighting the value added by market adjustments at probability extremes.
```

```python
## To add another layer to the analysis, we will do a favourite vs underdog calibration. 
##The point is to determine whether calibration differs if a player is expected to win or lose
```

```python
# Labelling favourites and underdogs
```

```python
# Copy evaluation dataframe to avoid mutation issues
eval_seg = eval_df.copy()

# Favourite if predicted probability >= 0.5
eval_seg["segment"] = np.where(
    eval_seg["predicted_proba"] >= 0.5,
    "Favourite",
    "Underdog"
)

eval_seg["segment"].value_counts()
```

```python
model_seg_calibration = (
    eval_seg
    .groupby(["segment", "proba_bin"])
    .agg(
        mean_predicted=("predicted_proba", "mean"),
        observed_win_rate=("actual_win", "mean"),
        count=("actual_win", "size")
    )
    .reset_index()
)

model_seg_calibration
```

```python
## Now do same for market 
```

```python
market_seg = market_df.copy()

market_seg["segment"] = np.where(
    market_seg["implied_proba"] >= 0.5,
    "Favourite",
    "Underdog"
)

market_seg["proba_bin"] = pd.cut(
    market_seg["implied_proba"],
    bins=np.linspace(0, 1, 11)
)

market_seg_calibration = (
    market_seg
    .groupby(["segment", "proba_bin"])
    .agg(
        mean_predicted=("implied_proba", "mean"),
        observed_win_rate=("win", "mean"),
        count=("win", "size")
    )
    .reset_index()
)

market_seg_calibration
```

```python
plt.figure()

# Model
for seg in ["Favourite", "Underdog"]:
    subset = model_seg_calibration[model_seg_calibration["segment"] == seg]
    plt.plot(
        subset["mean_predicted"],
        subset["observed_win_rate"],
        marker="o",
        label=f"Model – {seg}"
    )

# Market
for seg in ["Favourite", "Underdog"]:
    subset = market_seg_calibration[market_seg_calibration["segment"] == seg]
    plt.plot(
        subset["mean_predicted"],
        subset["observed_win_rate"],
        marker="x",
        linestyle="--",
        label=f"Market – {seg}"
    )

plt.plot([0, 1], [0, 1], linestyle=":", label="Perfect calibration")

plt.xlabel("Predicted Probability")
plt.ylabel("Observed Win Rate")
plt.title("Calibration by Market Expectation: Favourite vs Underdog")
plt.legend()
plt.show()
```

```python
### Favourite vs Underdog Calibration

#To examine whether calibration differs by market expectation, predicted probabilities were segmented into favourites (≥ 0.5) and underdogs (< 0.5).
#Calibration curves were constructed separately for both the data-driven model and bookmaker-implied probabilities.

#For favourites, both approaches exhibit strong calibration, with observed win rates closely tracking predicted probabilities.
#The betting market demonstrates particular strength at very high probability levels, consistent with the incorporation of
#expert and contextual information beyond historical performance metrics.

#For underdogs, calibration deteriorates for both the model and the market, reflecting greater outcome volatility. 
#However, the data-driven model shows slightly improved calibration in moderate underdog ranges, 
#suggesting that historical ranking and performance data may partially correct systematic market pessimism toward less-favoured players.

#Overall, the results indicate that while betting markets remain highly efficient—especially for strong favourites—simple probabilistic models
#built on public data can achieve competitive calibration and provide insight into areas of greater market uncertainty.
```
